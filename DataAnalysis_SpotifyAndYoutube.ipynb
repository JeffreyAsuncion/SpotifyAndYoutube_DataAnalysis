As for Views, Likes and Comment. df.info() shows that they are of data type float. For our analysis, we will need them to be of type int. Therefore, we will change the data types as needed.

```python
df['Views']=df['Views'].astype('int')
df['Likes']=df['Likes'].astype('int')
df['Comments']=df['Comments'].astype('int')
```

## Exploratory Data Analysis

```python
df.head(10)
```

### Determining the top 10 artists on the Spotify platform whose song have more views.

```python
view=df.groupby('Artist')['Views'].agg('sum').reset_index().sort_values('Views', ascending=False).head(10)
view
```

```python
view.set_index('Artist', inplace=True)
view.plot(kind='bar',color='g')
plt.xlabel('Artist')
plt.ylabel('Views')
plt.title('Top 10 Artist with higher views')
plt.xticks(rotation=75)
plt.show()
```

We can see here that Ed Sheeran was the most viewed artist with 15.5 billion views. With Cocomelon in second place with 14.6 billion views. And Katy Perry is in third place with 13.1 billion views. These figures highlight the popularity of these artists and their music, as well as the significant role of online video platforms in promoting and distributing music

### Determine the Top 10 artists with the Most Likes on Youtube

```python
df.groupby('Artist')['Likes'].agg('sum').reset_index().sort_values('Likes',ascending=False).head(10)
```

```python
ax=df.groupby('Artist').agg({'Likes':'sum','Stream':'sum'}).reset_index()
bx=ax.sort_values('Likes',ascending=False).head(10)
sns.barplot(data=bx,x='Likes', y='Artist')
plt.title('Top 10 Most Liked Artists')
plt.show()
```

As we can see here BTS is the most liked arts with 139 million likes. BLACKPINK is in second place with 132 million likes. And in third place, we have Charlie Puth with 86 million likes. 

## Determine the Top 10 Streamers on Spotify

```python
cx=ax.sort_values('Stream', ascending=False).head(10)
```

```python
df.groupby('Artist')['Stream'].agg('sum').reset_index().sort_values('Stream', ascending=False).head(10)
```

### As we can see Post Malone is the one in the first place streamer with 15.3 billion streams. Ed Sheeran is in second with 14.4 billion streams. In third place, we have Dua Lipa with 13.4 billion streams.

```python
sns.barplot(data=cx,y='Stream',x='Artist')
plt.xticks(rotation=80)
plt.title('Top 10 streamers')
plt.show()
```

## Who is the most commented artist on Youtube

```python
ax=df.groupby('Artist')['Comments'].sum().reset_index().sort_values(by='Comments', ascending=False).head(10)
ax
```

### We can see BTS in the first place position with 39.5 million comments.  BLACKPINK is in second place with 19.6 million comments. And in third place, we have Stray Kids with 8.5 million comments

```python
sns.barplot(data=ax,y='Artist', x='Comments')
plt.title('Top 10 most commented Artists')
plt.show()
```
